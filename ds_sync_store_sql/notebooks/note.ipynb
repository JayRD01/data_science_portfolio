```python
from dataclasses import asdict
from src.core.contracts.product_entry_contract import ProductEntryContract
from src.core.implementations.db_singleton_run import _DatabaseSingleton
from src.core.implementations.product_orm_run import ProductORM
from src.core.contracts.db_declarative_contract import Base
from src.core.implementations.table_writer_run import TableWriter
from src.core.implementations.table_reader_run import TableReader

# üß± Step 1: Initialize DB and create tables
db = _DatabaseSingleton()
Base.metadata.create_all(db.engine)

# ‚úÖ Step 2: Create validated contract object
product1 = ProductEntryContract(id=1, name="Laptop", price=1299.99)
product2 = ProductEntryContract(id=2, name="Mouse", price=20.00)

# üîÑ Step 3: Convert contract to ORM object
orm_product1 = ProductORM(**asdict(product1))
orm_product2 = ProductORM(**asdict(product2))

# üíæ Step 4: Use TableWriter to insert the product
writer = TableWriter(ProductORM, db)
writer.insert_row(orm_product1)
writer.insert_row(orm_product2)

# üëÄ Step 5: Use TableReader to display the table
reader = TableReader(ProductORM, db)
reader.show_table()
```
