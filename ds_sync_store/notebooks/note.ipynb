```python
from dataclasses import asdict
from src.core.contracts.product_entry_contract import ProductEntryContract
from src.core.implementations.db_singleton_run import _DatabaseSingleton
from src.core.implementations.product_orm_run import ProductORM
from src.core.contracts.db_declarative_contract import Base

# ðŸ§± Crear tablas si no existen
db = _DatabaseSingleton()
Base.metadata.create_all(db.engine)

# âœ… Crear instancia validada
product = ProductEntryContract(id=None, name="Laptop", price=1299.99)

# ðŸ”„ Convertir a ORM
# ðŸ”„ Convert dataclass -> dict y desempaquetar en el ORM
# Nota: asdict() solo convierte; no renombra ni mapea. Deben coincidir las claves.
orm_product = ProductORM(**asdict(product))  # => ProductORM(id=None, name="Laptop", price=1299.99)

# ðŸ’¾ Persistir
with db.get_session() as session:
    session.add(orm_product)
    print(f"âœ… Stored: {product.name} (${product.price})")
```
