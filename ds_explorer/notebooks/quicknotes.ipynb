```python
import fs
import pathlib
import os


os_sys = os.getcwd()
print(str(os_sys))

p = pathlib.Path(".").resolve()
print(str(p))

fs_ = fs.open_fs(".").getsyspath("/")
print(str(fs_))
```

```python
parent_os = os.path.abspath(os.path.join(os_sys,".."))
print(parent_os)

parent_pathlib = pathlib.Path.cwd().parent
print(parent_pathlib)

parent_fs = fs.open_fs("..").getsyspath("/")
print(parent_fs)
```

```python
new_os = os.path.abspath(os.path.join(parent_os, "notebooks"))
print(new_os)

new_pathlib = (parent_pathlib / "notebooks")
new_pathlib2 = pathlib.Path.joinpath(parent_pathlib, "notebooks")
print(f"{new_pathlib} \n{new_pathlib2} el siguiene es con fs")

new_fs = fs.path.join(parent_fs, "notebooks")
print(new_fs)

```

```python
import os
import tempfile

_tmp = tempfile.TemporaryDirectory()
rootpath = _tmp.name

os.makedirs(os.path.join(rootpath, 'subdir'))
with open(os.path.join(rootpath, 'hello.txt'), 'w', encoding='utf-8') as f:
    f.write('hi!')

print("Contenido:", os.listdir(rootpath))

_tmp.cleanup()
```
